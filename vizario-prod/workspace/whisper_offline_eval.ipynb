```python
import whisper
import os

# Number of test cases
NUM_TESTS = 5

# Path to the Whisper model checkpoint
WHISPER_MODEL_PATH = "base"  # Change if needed

# Paths to input and ground truth folders
INPUT_FOLDER = "./whisper_offfline_eval_inputs"
GT_FOLDER = "./whisper_offline_eval_gt"

# Load Whisper model
model = whisper.load_model(WHISPER_MODEL_PATH)

for i in range(1, NUM_TESTS + 1):
    audio_file = os.path.join(INPUT_FOLDER, f"whisper_eval_input{i}.m4a")
    gt_file = os.path.join(GT_FOLDER, f"whisper_gt{i}.txt")
    
    # Check files exist
    if not os.path.exists(audio_file):
        print(f"Audio file not found: {audio_file}")
        continue
    if not os.path.exists(gt_file):
        print(f"Ground truth file not found: {gt_file}")
        # continue

    # Transcribe audio
    result = model.transcribe(audio_file)
    transcript = result["text"].strip()

    # Read ground truth
    # with open(gt_file, "r", encoding="utf-8") as f:
    #     ground_truth = f.read().strip()
    
    # Print results
    print(f"===== Output {i} =====")
    print(transcript)
    print()
    print(f"===== Ground Truth {i} =====")
    # print(ground_truth)
    print()
```
